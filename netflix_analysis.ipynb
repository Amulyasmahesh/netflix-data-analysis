```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(df["type"].value_counts())
```

```python
print(df["country"].value_counts().head(10))
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

print(df['year_added'].value_counts())
```

```python
sns.countplot(x="type", data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Netflix Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Netflix Content")
plt.show()
```

## Objective
The objective of this project is to analyze Netflix content data to identify trends in movies, TV shows, genres, and countries.

## Dataset
The dataset contains information about Netflix titles including:
- Type (Movie or TV Show)
- Country
- Genre
- Release Year
- Duration

## Analysis Questions

1. What is the distribution of Movies vs TV Shows on Netflix?
2. Which countries produce the most Netflix content?
3. How has Netflix content grown over the years?
4. What are the most popular genres on Netflix?
5. What are the longest movies available?

## Key Insights

• Netflix hosts significantly more movies than TV shows.

• The United States produces the highest number of Netflix titles.

• Netflix content production increased rapidly after 2016.

• International Movies and Dramas are among the most common genres.

• The Netflix library has expanded significantly over the last decade.
